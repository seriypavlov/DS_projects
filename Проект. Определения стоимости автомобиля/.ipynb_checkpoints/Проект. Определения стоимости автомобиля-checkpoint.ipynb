# Проект. Определения стоимости автомобиля

## Глава 0. Входные данные

Сервис по продаже автомобилей с пробегом «Не бит, не крашен» разрабатывает приложение для привлечения новых клиентов. В нём можно быстро узнать рыночную стоимость своего автомобиля. В вашем распоряжении исторические данные: технические характеристики, комплектации и цены автомобилей. Нужно построить модель для определения стоимости. 

Заказчику важны:

- качество предсказания;
- скорость предсказания;
- время обучения.

#### Путь к файлу:
- /datasets/autos.csv

#### Описание данных:   
Признаки:  
- `DateCrawled` — дата скачивания анкеты из базы  
- `VehicleType` — тип автомобильного кузова  
- `RegistrationYear` — год регистрации автомобиля  
- `Gearbox` — тип коробки передач

Целевой признак:
- `Price` — цена (евро)

```python
df = df.drop(['DateCrawled', 'DateCreated', 'LastSeen'], axis=1)
df
```

Приведем столбец `date` к формату даты:

Проверим признаки `RegistrationMonth` и `RegistrationYear` на уникальность:

```python
df.RegistrationMonth.unique()
```

```python
sorted(df.RegistrationYear.unique())
```

Т.к. 0 недопустимое значение признака `RegistrationMonth`, и, вероятно, возникло из за того, что владелец не указал месяц регистрации авто, заменим на Январь(т.е. 1):

```python
df.RegistrationMonth = df.RegistrationMonth.replace(0, 1, inplace=False)
df.RegistrationMonth.unique()
```

Выведем df с признаком `RegistrationYear`, у которого год указан аномальным значением:

```python
invalid_year = df[(df.RegistrationYear < 1910) | (df.RegistrationYear > 2019)]
invalid_year
```

Количество объектов 171, что составляет < 0.01 % всех объектов. Удалим объекты с аномальными значениями по индексам.

```python
df.drop(invalid_year.index, inplace=True)
```

Сохраним количественные признаки в переменную **numeric_feature**:

```python
numeric_feature = set(('Price', 'Power', 'Kilometer'))
```

Напишем функцию **numeric_feature_distribution**, строящую график распределения количественного признака:

```python
def numeric_feature_distribution(column):
    sns.set()
    fig, ax = plt.subplots(figsize=(12, 5))
    sns.distplot(df[column])
    ax.set_title('Распределение по ' + column)
    plt.ylabel('Плотность частоты');
```

Построим графики количественных признаков:

```python
[numeric_feature_distribution(item) for item in numeric_feature];
```

Признак `Kilometer` относиться больше к категориальным признакам, удалим из переменной:

```python
numeric_feature.discard('Kilometer')
numeric_feature
```

Посмотрим на распределения в признаке `Price`:

```python
df.Price.describe(percentiles=[.05, .1, .15, .25, .5, .75, .95])
```

Ограничим распределение 5% и 95% квантилями, исключив аномальные значения:  

```python
invalid_price = df[(df.Price < df.Price.quantile(.05)) | (df.Price > df.Price.quantile(.95))]
invalid_price.shape
```

```python
df.drop(invalid_price.index, inplace=True)
```

Посмотрим на распределения в признаке `Power`:

```python
df.Power.describe(percentiles=[.05, .1, .15, .25, .5, .75, .95])
```

Мощность двигателя ниже 15% и выше 95% квантилей, является аномальным. Удалим их из респределения:

```python
invalid_power = df[(df.Power < df.Power.quantile(.15)) | (df.Power > df.Power.quantile(.95))]
invalid_power.shape
```

```python
df.drop(invalid_power.index, inplace=True)
```

Снова построим графики количественных признаков:

```python
[numeric_feature_distribution(item) for item in numeric_feature];
```

Сохраним категориальные признаки в переменную **categorial_features**:

```python
categorial_features = set(df.columns) - numeric_feature
categorial_features
```

Посмотрим на распределения в категориальных признаках методом `.value_counts()`:

```python
for item in categorial_features:
    print('{: ^20}'.format(item))
    print(df[item].value_counts())
    print()
```

Напишем функцию **categorial_feature_distribution**, строящую график распределения категориального признака:

```python
def categorial_feature_distribution(column):
    fig, ax = plt.subplots(figsize=(12, 5))
    plt.xticks(rotation=90)
    sns.countplot(x=column, data=df, palette='coolwarm')
    ax.set_title('Распределение по ' + column)
    plt.ylabel('Количество');
```

Построим распределения тех признаков, что легко визуализировать(кроме `Model` и `PostalCode`):

```python
categorial_feature_for_plot = categorial_features - set(('Model', 'PostalCode'))
categorial_feature_for_plot
```

Построим графики категориальных признаков:

```python
[categorial_feature_distribution(item) for item in categorial_feature_for_plot];
```

По диаграммам видно, что `NumberOfPictures` принимает только одно значение. Так же удалим `PostalCode` как не несущих особой ценности для модели обучения, а только усложняющий обучение:

```python
df = df.drop(['NumberOfPictures', 'PostalCode'], axis=1)
df
```

Оставшиеся пропуски сложно восстановить без дополнительной информации. Оставим без изменения.

```python
df.isna().sum()
```

Чтобы преобразовать категориальные признаки в численные поможет техника прямого кодирования (One-Hot Encoding, OHE).
Для прямого кодирования в библиотеке pandas есть функция `pd.get_dummies()`  
чтобы избежать дамми-ловушки, используем аргумент `drop_first=True`

```python
df_ohe = pd.get_dummies(df, drop_first=True)
df_ohe
```

Разделим датафрейм на 2 части: обучающую и тестовую выбороки в пропорциях 4:1, используя функцию `train_test_split` из библиетеки `sklearn.model_selection`  

```python
df_train, df_test = train_test_split(df_ohe, test_size=0.2, random_state=12345)
```

Разделим выборки на `features` -  признаки и `target` — целевой признак.

```python
train_features = df_train.drop(['Price'], axis=1)
train_target = df_train['Price']
test_features = df_test.drop(['Price'], axis=1)
test_target = df_test['Price']
```

Для того, чтобы отмаштабировать признаки, воспользуемся одним из методов масштабирования — стандартизации данных.  
В sklearn есть отдельная структура для стандартизации данных — **StandardScaler**  
Преобразуем обучающую,валидационную и тестовую выборки функцией `transform()`.

```python
numeric = train_features.iloc[:, :4].columns
scaler = StandardScaler()
scaler.fit(train_features[numeric])

train_features[numeric] = scaler.transform(train_features[numeric])
test_features[numeric] = scaler.transform(test_features[numeric])
```

### Выводы по главе 1:

> В процессе загрузки и обработки данных обнаружили:  
- В данных присутствуют неинформативные признаки для модели, эти признаки нужны для сервиса, но не несут ценности для обучения модели и увеличивают время обучения, поэтому были удалены.  
- Дубликаты и аномальные значения были удалены так же. Вероятно, что первые появились в процессе выгрузки данных, вторые человеческий фактор(пользователи неверно указывают данные). Пропуски сложно восстановить по имющимся данным, поэтому оставлены как есть. Модели сами учтут их в процессе работы. 

## Глава 2. Обучение моделей

Импортируем **make_scorer** из **sklearn.metrics** и напишем функцию потерь и оценки качества моделей **RMSE**:

```python
from sklearn.metrics import make_scorer
```

```python
def RMSE(target, predict):
    RMSE = mean_squared_error(target, predict) ** 0.5
    return RMSE

score = make_scorer(RMSE, greater_is_better=False)
```

Для подбора гиперпараметров импортируем **GridSearchCV**, для кросс-валидации при подборе гиперпараметров **ShuffleSplit**:

```python
cv = ShuffleSplit(n_splits=3, test_size=0.2, random_state=12345)
```
